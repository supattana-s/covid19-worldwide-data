```python
import pandas as pd
```

```python
data = pd.read_csv('./Documents/data/covid_case_woldwide/covid_worldwide.csv')
```

```python
columns = data.columns
```

```python
changing_columns = []
```

```python
for idx in range(0, len(columns)):
    if columns[idx] != 'Country':
        changing_columns.append(columns[idx])
```

```python
changing_columns
```

```python
to_change_values = data[changing_columns]
```

```python
changed = to_change_values.apply(lambda x: pd.to_numeric(x.str.replace(',', '').fillna(0), downcast='integer') if x.dtype == 'object' else x)
# changed = to_change.apply(lambda x: print(x.dtype))
```

```python
changed
```

```python
country_only = data[['Serial Number', 'Country']]
```

```python
country_only
```

```python
cleaned_data = country_only.merge(changed, how='inner', on='Serial Number')
```

```python
cleaned_data
```

```python
cleaned_data.to_csv('./Documents/data/covid_case_woldwide/covid19_worldwide_cleaned.csv', index=False)
```
